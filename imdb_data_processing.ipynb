```python
#remove whitespace
MPAA_df['MPAA_Rating'] = MPAA_df['MPAA_Rating'].str.strip()
```

```python
#check how many valid ratings we have
starts_with_tv = 0
invalid_rating = 0

for rating in MPAA_df['MPAA_Rating']:
    if rating.startswith('TV'):
        starts_with_tv += 1
    else:
        invalid_rating += 1
        
print(starts_with_tv, invalid_rating)
```

```python
def remove_white_space(df):
    for column in df:
        df[column] = df[column].str.strip()
    return df
```

```python
remove_white_space(MPAA_df)
```

```python
#pickle the data
MPAA_df.to_pickle('MPAA_df.pkl')
```

```python
#make a copy just in case
MPAA_filtered = MPAA_df.copy(deep = True)
```

```python
#if the rating isn't valid, replace it with None
for idx, rating in enumerate(MPAA_filtered['MPAA_Rating']):
    if not rating.startswith('TV'):
        MPAA_filtered['MPAA_Rating'][idx] = None
```

### Find the production networks that are big and then create a binary column that says whether or not a show was produced by one of these large companies

```python
for i in range(30):
    print(Filtered_Production_df['Title'][i], '|',Filtered_Production_df['Production_Co'][i])
```

```python
#grab the large networks i see
big_networks = ['Home Box Office', 'HBO', 'Sony Pictures Television', 'Netflix', 
                'Paramount', 'DC Comics', 'FX Productions', 'Amazon', 'ABC', 'Disney', 
                'Marvel', 'Bad Robot', 'Warner Bros', '20th Century Fox', 'DC Entertainment', 
                'Showtime']
```

```python
#create a binary list based on whether or not a large prod comp is in the row
large_prod = [0]*502

for i in range(len(Filtered_Production_df)):
    for prod_co in big_networks:
        
        if (large_prod[i] == 0) and (prod_co in Filtered_Production_df['Production_Co'][i]):
            large_prod[i] = 1
            
            
print(large_prod[:30])
```

```python
len(large_prod)
```

```python
#create a new column
Filtered_Production_df['Large_prod_co'] = large_prod

Filtered_Production_df.head(15)
```
